# Leading Causes of Death in United States of America
This study analyses the leading causes of death in the United States of
America between 1999 and 2017
This case study will try to answer the following questions:
- What is the total number of records in the dataset?
- What were the causes of death in this data set?
- What was the total number of deaths in the United
- States from 1999 to 2017?
- What is the number of deaths per each year from 1999
to 2017?
- Which ten states had the highest number of deaths
overall?
- What were the top causes of deaths in the United States
during this period?

```python
# Load packages
import pandas as pd

# Load data
data = pd.read_csv("NCHS_-_Leading_Causes_of_Death__United_States.csv")
data.head(3)
```

```python
# Checks
data.shape # (10868 rows and 6 columns)
data = data.dropna() # Remove all rows with na cases
data.shape # Approximately 10,868 death cases were recorded in different U.S. states
```

```python
data["Year"].max() # year range 1999 to 2017
data["Year"].min()
```

```python
# Data Cleaning
# Unique Death Causes in the United States
causes = data["Cause Name"].unique()
causes # Remove 'All causes' from the Cause Name column
data = data[data["Cause Name"] !="All Causes"]
causes = data["Cause Name"].unique()
causes
```

```python
len(causes)
```

```python
# Unique States in the Study
state = data["State"].unique()
state
```

```python
len(state) #  52 states included in the study
```

```python
# Total number of deaths in the United States from 1999 to 2017
data["Deaths"].sum()
# The total number of deaths during the given period is 168,018,304.
```

```python
# What is the number of deaths for each year from 1999 to 2017?
deaths_per_year = data.groupby(["Year"]).sum()
deaths_per_year
```

```python
import matplotlib.pyplot as plt  # Importing the pyplot module

# Graph deaths per year
deaths_per_year["Deaths"].plot(title = "Deaths per year \n 1999-2017")

# Set x-ticks to be integers
plt.xticks(ticks=range(1999, 2018, 2))

plt.show()

# The number of deaths declined between 2002 and 2009. 
# Then there was a continuous growth in the number of deaths from 2010 to 2013. 
# Finally, there was a sharp increase in the number of deaths in 2013and 2014
```

```python
# Top Ten States with the Highest Number of Deaths in the United States
data1 = data[data["State"] != "United States"]
dataset2 = data1.groupby("State").sum()
dataset2.sort_values("Deaths", ascending = False, inplace = True)
dataset2.head(10)
```

```python
dataset2["Deaths"].head(10).plot.bar(title = "Top ten states with the highest death number \n 1999-2017")
# California had the highest number of deaths in the United States, with Florida coming in second.
```

```python
# Top causes of deaths in the United States during this period?
dataset1 = data[data["Cause Name"] !="All Causes"]
dataset2 = dataset1.groupby("Cause Name").sum()
dataset2.sort_values("Deaths", ascending = False, inplace = True)
dataset2.head(10)
```

```python
dataset2["Deaths"].head(10).plot.bar(title = "Top ten causes of death in USA \n 1999 - 2015")
```
